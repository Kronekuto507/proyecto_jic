```python
import random
main_path_dataset = "..\datasets\PSL_Main"
paths_in_dataset = [os.path.join(main_path_dataset,letter) for letter in os.listdir(main_path_dataset)]
diccionario = {path_name : os.listdir(path_name) for path_name in paths_in_dataset}
random_values_dataset = [path + "\\" + lista[random.randint(0,int(len(lista)))] for path,lista in diccionario.items()]
```

```python
opencv_figures = [cv2.resize(cv2.cvtColor(cv2.imread(image),cv2.COLOR_BGR2RGB), (120,120)) for image in random_values_dataset] 
len(opencv_figures)
titles = ['A','B','C','CH','D','E','F','G','H','I','K','L','M','N','O','P','Q','R','S','T','U','V','W','X','Y']
```

```python
fig,axes = plt.subplots(2,12,figsize = (30,5))
axes = axes.flatten()
figsize=(255,255)
for ax,title,img in zip(axes,titles,opencv_figures):
    ax.imshow(img)
    ax.set_title(title[-1])
    ax.axis('off')
plt.show()
```
