```python
import os
import shutil
from pathlib import Path
```

```python
ruta_destino = Path("..\\datasets\\nz_dataset")
```

```python
if not os.path.exists(ruta_destino):
    os.makedirs(ruta_destino)
```

```python
path_subdirs_letter = Path("..\\datasets\\ASL_Dataset_Combination")
```

```python
subdirs_name = os.listdir(path_subdirs_letter)
```

```python
for name in subdirs_name:
    path = os.path.join(ruta_destino,name)
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
ruta_vieja = Path("..\\datasets\\datasets_individuales\\datasets_tar\\nz_dataset")
```

```python
for subdir in os.listdir(ruta_destino):
    for file in os.listdir(ruta_vieja):
        print(file[0])
        if file[0] == subdir:
            old_path = os.path.join(ruta_vieja,file)
            new_path = os.path.join(ruta_destino,subdir,file)
            shutil.copy2(old_path,new_path)
    
```
