```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import PIL
import mediapipe as mp
from pathlib import Path
```

```python
import math
def detectar_mano_bbox(imagen):
    #Iniciar deteccion de manos
    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands()
    #Convertir imagen a formato de cv2
    imagen = cv2.imread(imagen, cv2.IMREAD_COLOR)
    results = hands.process(imagen)
    x_min = y_min = float('inf')
    x_max = y_max = float('-inf')
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            for lm in hand_landmarks.landmark:
                x, y = int(lm.x * imagen.shape[1]), int(lm.y * imagen.shape[0])
                x_min = min(x_min, x)
                y_min = min(y_min, y)
                x_max = max(x_max, x)
                y_max = max(y_max, y)
        # Dibujar bbox de la mano
        cv2.rectangle(imagen, (x_min - 35, y_min - 35), (x_max + 35, y_max + 35), (255, 255, 0), 2)
        return imagen,[(x_min - 35, y_min - 35),(x_max + 35, y_max + 35)]
    return imagen, None
        
```
